```python
# Import all datasets except for simple_submission

import numpy as np
import pandas as pd

aisles = pd.read_csv('instacart-market-basket-analysis/aisles.csv')
departments = pd.read_csv('instacart-market-basket-analysis/departments.csv')
order_products_prior = pd.read_csv('instacart-market-basket-analysis/order_products__prior.csv')
order_products_train = pd.read_csv('instacart-market-basket-analysis/order_products__train.csv')
orders = pd.read_csv('instacart-market-basket-analysis/orders.csv')
products = pd.read_csv('instacart-market-basket-analysis/products.csv')
```

```python
# Check to make sure each aisle is unique to a specific department

product_categories = pd.merge(left=products, right=aisles, how='left', left_on='aisle_id', right_on='aisle_id')
product_categories = pd.merge(left=product_categories, right=departments, how='left', left_on='department_id', right_on='department_id')

def check_aisle_department(x, match):
    if x['aisle_id'] not in match.keys():
        match[x['aisle_id']] = x['department_id']
    else:
        if match[x['aisle_id']] != x['department_id']:
            print(False)

match = {}

product_categories.apply(check_aisle_department, args = (match,), axis = 1)
    
product_categories.sort_values(['department_id','aisle_id'])
```

```python
# Code to combine order csv's into a single usable order dataset

# 1. Remove any orders assigned to the test dataset
new_ds = orders.loc[(orders['eval_set'] == 'prior') | (orders['eval_set'] == 'train')]
print("Updated unique list of evaluation sets:", new_ds['eval_set'].unique())

# 2. Combine the prior and train sets
prior_and_train = order_products_prior.append(order_products_train, ignore_index=True)
print("No change in size of combined datasets:", 
      len(prior_and_train) == (len(order_products_prior)+len(order_products_train)))

# 3. Check for any missing values in new prior_and_train and orders datasets
print("Missing values in new prior_and_train:", prior_and_train.isna().any().any())
print("Missing values in orders:", orders.isna().any().any())

print(orders.loc[:, orders.isnull().any()].columns) # Only the days_since_prior_order has NA, which we can ignore

# 4. Combine orders and prior_and_train using a leftjoin

new_ds = pd.merge(left=prior_and_train, right=orders, how='left', left_on='order_id', right_on='order_id')

# 5. Delete the eval_set column

new_ds.drop('eval_set', axis = 1, inplace=True)

# 6. Check for duplicate values within new_ds by looking at the order_id, add_to_cart_order pairing

new_ds.duplicated(subset=['order_id', 'add_to_cart_order']).unique()
```

```python
new_ds
```

```python
# Conduct checks to make sure datasets joined properly
print("Size is the same:", len(prior_and_train) == len(new_ds))
print("Which columns have NA values:", new_ds.loc[:, new_ds.isnull().any()].columns)
```

```python
# Add in the product csv

# 1. Check for missing values in any of the products, aisles, or departments datasets
print("Missing values in products:", products.isna().any().any())
print("Missing values in aisles:", aisles.isna().any().any())
print("Missing values in prior_and_train:", departments.isna().any().any())

# 2. Merge products with main dataset, new_ds using a leftjoin

new_ds_products = pd.merge(left=new_ds, right=products, how='left', left_on='product_id', right_on='product_id')
new_ds_products
```

```python
# Add in aisles and departments csv's

# 1. Merge aisles and departments with main dataset, new_ds_products using a leftjoin

new_ds_prod_aisles = pd.merge(left=new_ds_products, right=aisles, how='left', 
                              left_on='aisle_id', right_on='aisle_id')
new_ds_complete = pd.merge(left=new_ds_prod_aisles, right=departments, how='left', 
                              left_on='department_id', right_on='department_id')
new_ds_complete
```

```python
# Rearrange the column order

new_ds_complete = new_ds_complete[['user_id','order_number','order_id','order_dow','order_hour_of_day',
                                   'days_since_prior_order','product_id','product_name','add_to_cart_order',
                                   'reordered', 'aisle_id', 'aisle', 'department_id', 'department']]
new_ds_complete
```

```python
# Sort the dataset by user_id, order_number, and add_to_cart_order in that priority
new_ds_complete.sort_values(['user_id', 'order_number', 'add_to_cart_order'], inplace=True)
```

```python
new_ds_complete
```

```python
# Create a produce only dataset
produce_ds = new_ds_complete[(new_ds_complete['department'] == 'produce')]

# Drop the department_id and name because all rows share the same values
produce_ds.drop(['department_id','department'], axis = 1, inplace=True)
```
